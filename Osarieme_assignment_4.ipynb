````python
class NumericTransformers:
    """This class is used to apply sigmoid and cube root transformation to the input data(Vector)"""
    
    def __init__ (self, *, x_vector: np.ndarray) -> None:
        self.x_vector = x_vector.copy()
        self.x_mean = np.mean(x_vector)
        self.std = np.std(x_vector)
        self.x_min = np.min(x_vector)
        self.x_max = np.max(x_vector)
        
    def __repr__(self) -> str:
        """This returns the string representation."""
        return (
            f"{self.__class__.__name__}(mean = {round(self.x_mean, 3)}, "
            f"std = {round(self.std, 3)},"
            f"min = {self.x_min},"
            f"max = {self.x_max})"
        )      
````
